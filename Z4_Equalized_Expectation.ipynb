# Equalized Expectation

Solve the following games:
1. Reduce by dominance, if possible, and as far as possible.
2. Equalize expectations if game to reduces to $3\times 3$ or $4\times 4$, but no further.

## Example 1: Reduction by Dominance
**Solve the game.**
$$\begin{array}{cc}&\textbf{Colin}\\\textbf{Rose}&\begin{array}{r|rrrr}&A&B&C\\ \hline A&6&0&6\\B&3&-2&-2\\C&8&3&3\end{array}\end{array}$$

## Example 2: Reduction by Dominance
Solve the game.
$$\begin{array}{cc}&\textbf{Colin}\\\textbf{Rose}&\begin{array}{r|rrr}&A&B&C\\ \hline A&6&3&4\\B&1&11&6\\C&0&11&8\end{array}\end{array}$$

### Example 3: A $3\times 3$ with no Reduction
$$\begin{array}{cc}&\textbf{Colin}\\\textbf{Rose}&\begin{array}{r|rrr}&A&B&C\\ \hline A&12&2&2\\B&-2&-3&12\\C&0&7&2\end{array}\end{array}$$
