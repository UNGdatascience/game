# Expected Value

In a matrix game, given that one player knows through experiencing repeated play (or other means) the strategy mixture being played by the opponent, she can calculate the \textbf{\textcolor{blue}{expected value}} of each of her strategies.

## Definition

Given a set $X$ of $n\in\mathbb{N}$ outcomes and $n$ associate probabilities, the <span style="color:red;"><b>expected value</b></span> of $X$ is given as follows:

$$E(X) =\sum x_ip(x_i)$$

The formula above is well know from probability theory. Over many repititions of the game, the expected value of a strategy corresponds to the payoff a player earns when playing that strategy every time. The idea transitions in statistics into the mean or **average**.

### Example 1
Find the expected value of strategies $\textbf{Rose A}$ and $\textbf{Rose B}$ in the following game provided that Colin plays $\vec c = (0,0.75,0.25)$.
$$\begin{array}{cc}&\textbf{Colin}\\\textbf{Rose}&\begin{array}{r|rrr}&A&B&C\\ \hline A&8&0&-3\\B&-4&-4&9\end{array}\end{array}$$
We notice immediately the dominance in $\textbf{Colin A}$ and $\textbf{Colin B}$ where
$$\begin{align}
  0&\leq 8\\
  -4&\leq-4 \implies\\
  \textbf{Colin B}&\leq \textbf{Colin A}
\end{align}$$
Remember, <span style="color:red;"><b>Colin is minimizing</span></b>. Thus, we are finding the expected values for Rose in the reduced game:
$$\begin{array}{cc}&\textbf{Colin}\\\textbf{Rose}&\begin{array}{r|rr}&B&C\\ \hline A&0&-3\\B&-4&9\end{array}\end{array}$$
