# Module Z2: Dominance

Dominated Strategy
: A strategy $S$ dominates a strategy $T$ if every entry for $S$ is greater than or equal to the corresponding entry for $T$. 

If this is the case, we say that $T$ is dominated by $S$ and write:

$$S\geq T$$

## Examples

1. Solve the game using reduction by dominance.
$$\begin{array}{cc}&\textbf{Colin}\\\textbf{Rose}&\begin{array}{r|rr}&A&B\\ \hline A&7&10\\B&-4&11\end{array}\end{array}$$

2. Solve the game using reduction by dominance.

$$\begin{array}{cc}&\textbf{Colin}\\\textbf{Rose}&\begin{array}{r|rrrr}&A&B&C&D\\ \hline A&8&7&-2&1\\B&8&7&4&5\\C&0&-2&12&-1\end{array}\end{array}$$

3. Solve the game using reduction by dominance.

$$\begin{array}{cc}&\textbf{Colin}\\\textbf{Rose}&\begin{array}{r|rrrr}&A&B&C&D\\ \hline A&7&-1&5&-3\\B&8&11&12&12\\C&3&-2&-3&-4\\D&-3&10&3&9\end{array}\end{array}$$
